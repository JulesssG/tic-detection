```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from ast import literal_eval

from video_loader import *
from utils import *
```

```python
video = VideoLoader('data/trial1_2020_12_16_1280x720.mp4')
meta = pd.read_excel('data/trial1_2020_12_16_transcriptions.xls')
meta
```

```python
header_index = np.where(meta.iloc[:50,0].str.contains('CSV_HEADER'))[0][0]
meta.columns = meta.iloc[header_index].str.strip('# CSV_HEADER = ').values
meta = meta.iloc[header_index+1:]
meta
```

#### Process metadata

```python
file = meta.file_list.values[0].strip("\"''[]")
file = 'data/trial1_2020_12_16_1280x720.mp4'

fps = int(cv2.VideoCapture(file).get(cv2.CAP_PROP_FPS))

# https://stackoverflow.com/questions/43600878/merging-overlapping-intervals
def merge_intervals(intervals):
    starts = intervals[:,0]
    ends = np.maximum.accumulate(intervals[:,1])
    valid = np.zeros(len(intervals) + 1, dtype=np.bool)
    valid[0] = True
    valid[-1] = True
    valid[1:-1] = starts[1:] >= ends[:-1]
    return np.vstack((starts[:][valid[:-1]], ends[:][valid[1:]])).T


normal_intervals = np.array(meta.temporal_coordinates.apply(literal_eval).to_list())
normal_intervals = merge_intervals(np.sort(normal_intervals, axis=0))

normal_intervals_f = (normal_intervals*fps).astype(int)
start_frame = normal_intervals_f.ravel()[0]
end_frame = normal_intervals_f.ravel()[-1]
normal_intervals_f = normal_intervals_f - start_frame
nframes = end_frame-start_frame+1
mask_no_tic = np.full(nframes, False)
for a, b in normal_intervals_f:
    mask_no_tic[a:b] = True
print(start_frame, end_frame, nframes,len(mask_no_tic))
tic_masked = np.ma.masked_array([0]*nframes, mask_no_tic)

tic_intervals_f = np.roll(normal_intervals_f, -1)[:-1]

vid_args = {'torch': False,
           'gray': True,
           'scale': (512, 512),
           'sample_shape': (512**2,)}

ncomp = 10
```

#### Define the videos

```python
normal_videos = [VideoLoader(file, start_frame=a, duration_frames=b-a+1, **vid_args) for a, b in normal_intervals_f]
tic_videos = [VideoLoader(file, start_frame=a, duration_frames=b-a+1, **vid_args) for a, b in tic_intervals_f]
full_video = VideoLoader(file, start_frame=start_frame, duration_frames=nframes, **vid_args)
```

#### Fit pca to frames without tic

```python
all_normal_frames = []
for video in normal_videos:
    all_normal_frames.append(video.get_all_frames())
all_normal_frames = np.vstack(all_normal_frames)
```

```python
normal_pca = custom_pca(ncomp)
normal_pca.fit(all_normal_frames)
```

#### Encode the video and compute transition matrix for all frames

```python
all_frames_enc = []
for frames in full_video:
    all_frames_enc.append(normal_pca.encode(frames)[0])
all_frames_enc = np.vstack(all_frames_enc)

A = np.linalg.pinv(all_frames_enc[:-1])@all_frames_enc[1:]
```

#### Reconstruct all frames and compute the reconstruction error

```python
all_frames_rec = []
for frames in np.array_split(all_frames_enc, full_video.duration_frames//full_video.batch_size, axis=0):
    all_frames_rec.append(normal_pca.decode(frames))
all_frames_rec = np.vstack(all_frames_rec)[:-1]
```

```python
errors = np.mean((all_frames_rec - full_video.get_all_frames()[1:])**2, axis=1)
```

#### Plot the results

```python
plt.figure(figsize=(22,10))
plt.title('Reconstruction error using general dynamics and PCA model from frames without tic')
plt.plot(errors, label='Reconstruction error')
plt.plot(tic_masked, label='Tic occurences')
plt.xlabel('Mean per pixel error')
plt.ylabel('Frame number')
_ = plt.legend()
```
